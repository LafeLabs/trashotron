# PLOTS OF SLIME DATA



```python
import requests
import json
import matplotlib.pyplot as plt
# Fetch the JSON data from the URL
url = "https://raw.githubusercontent.com/LafeLabs/trashotron/main/slimedata/2023_4_22/files.txt"
response = requests.get(url)
files = json.loads(response.text)
```

```python
base = "https://raw.githubusercontent.com/LafeLabs/trashotron/main/slimedata/2023_4_22/"
fileurls = []
for file in files:
    print(base + file)
    fileurls.append(base + file)
    
dataarray = []
for fileurl in fileurls:
    response = requests.get(fileurl)
    data = json.loads(response.text)
    dataarray.append(data)
    
```

```python
data = dataarray[7]
```

```python
import numpy as np
spectra = []
for obj in data:
    spectrum = np.array(obj["spectrum"])
    spectra.append(spectrum)

    
spectra_2d = np.vstack(spectra)
fig, ax = plt.subplots(figsize=(14, 8))
im = ax.imshow(spectra_2d.T[:][:-900], aspect="auto", origin="lower")
ax.set_xlabel("Index of top-level array")
ax.set_ylabel("Index of spectrum array")
ax.set_title("Spectrum data")
fig.colorbar(im)
plt.show()

```

```python
z_values = [obj['z'] for obj in data]

# Plot the 'z' values
plt.plot(z_values)
plt.xlabel("Index of top-level array")
plt.ylabel("Value of 'z'")
plt.title("Values of 'z'")
plt.show()
```

```python
z_values = [obj['y'] for obj in dataarray[6]]

# Plot the 'z' values
plt.plot(z_values)
plt.xlabel("Index of top-level array")
plt.ylabel("Value of 'y'")
plt.title("Values of 'y'")
plt.show()
```

```python
trajectory = []

for scan in dataarray:
    for point in scan:
        pointdata = json.loads("{}")
        pointdata['t'] = point['t']
        pointdata['x'] = point['x']
        pointdata['y'] = point['y']
        pointdata['z'] = point['z']
        pointdata['sum'] = int(np.sum(point['spectrum']))
        pointdata['maxfreqindex'] = int(np.argmax(point['spectrum']))
        trajectory.append(pointdata)
```

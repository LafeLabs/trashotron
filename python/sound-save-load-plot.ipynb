```python
import sounddevice as sd
from scipy.io.wavfile import write

fs = 44100  # Sample rate
seconds = 10  # Duration of recording

# Record audio
print("Recording...")
myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=2)
sd.wait()  # Wait until recording is finished
print("Finished recording")

# Save audio as WAV file
write("myaudio3.wav", fs, myrecording)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

# Load the WAV file
fs, data = wavfile.read('myaudio3.wav')

# Flatten the data array
data = data.ravel()

# Set up the spectrogram parameters
window_size = 1024
hop_size = 512
n_fft = window_size

# Compute the spectrogram
spec, freqs, times, _ = plt.specgram(data, NFFT=n_fft, Fs=fs, window=np.hamming(window_size), noverlap=hop_size)

# Extract first row of spectrogram
spectrogram_row = spec[0, :]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

# Load the WAV file
fs, data = wavfile.read('myaudio.wav')

# Flatten the data array
data = data.ravel()

# Set up the spectrogram parameters
window_size = 1024
hop_size = 512
n_fft = window_size

f = plt.figure()
f.set_figwidth(10)
f.set_figheight(10)
plt.xlabel('t [s]')
plt.ylabel('f [Hz]')
plt.title('fft(slimezistor)')

# Compute the spectrogram
spec, freqs, times, _ = plt.specgram(data, NFFT=n_fft, Fs=fs, window=np.hamming(window_size), noverlap=hop_size)

# Extract first row of spectrogram
spectrogram_row = spec[0, :]
```

```python
time_in_seconds = 16.5
time_in_frames = int(time_in_seconds * fs / hop_size)
spectrum = spec[:, time_in_frames]
freqs = np.fft.rfftfreq(n_fft, 1/fs)
plt.plot(freqs, spectrum,"+")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
